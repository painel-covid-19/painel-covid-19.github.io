```python
def get_last_update_date(data):
    last_updates = {}
    for index, row in enumerate(data["new_confirmed"]):
        if row >= 1:
            last_updates["city"] = data["city"].loc[0]
            last_updates["date"] = data.loc[index, "date"]
            break
    return last_updates
```

```python
import os
import csv

directory = "dt"

if not os.path.exists(directory):
    os.makedirs(directory)

with open(f"{directory}/last_updated_dates.csv", "w", encoding="utf-8") as csv_file:
    csv_writer= csv.writer(csv_file)
    csv_writer.writerow(["city", "dates"])
    for row in last_updated_date:
        city = row["city"]
        date = row["date"]
        csv_writer.writerow([
            city,
            date
        ])
```
